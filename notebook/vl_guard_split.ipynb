```python
import json
```

```python
safe = [] # retain
unsafe = [] 

with open('../VLGuard/data/train.json') as fout:
    # print(len(json.load(fout)))
    all_samples = json.load(fout)
    for sample in all_samples:
        if sample['safe']:
            safe.append(sample)
        else:
            unsafe.append(sample)
print(f"[Safe], {len(safe)}")
print(f"[Unsafe], {len(unsafe)}")
```

```python
with open('../VLGuard/data/retain_data.json', 'w') as fin:
    json.dump(safe, fin, indent=4)
with open('../VLGuard/data/forget_data.json', 'w') as fin:
    json.dump(unsafe, fin, indent=4)
```

```python
with open('../VLGuard/data/retain_data.json', 'r') as fout:
    retain_data = json.load(fout)
    reformulate_data = []
    for sample in retain_data:
        reformulate_data.append(
            {
                'id': f"{sample['id']}_safe_inst",
                'image': sample['image'],
                'conversations': [
                    {'from': 'human', 'value': sample['instr-resp'][0]['safe_instruction']},
                    {'from': 'gpt', 'value': sample['instr-resp'][0]['response']}
                ]
            }
        )
        reformulate_data.append(
            {
                'id': f"{sample['id']}_unsafe_inst",
                'image': sample['image'],
                'conversations': [
                    {'from': 'human', 'value': sample['instr-resp'][1]['unsafe_instruction']},
                    {'from': 'gpt', 'value': sample['instr-resp'][1]['response']}
                ]
            }
        )
with open('../VLGuard/data/retain_data.json', 'w') as fin:
    json.dump(reformulate_data, fin, indent=4)
```

```python
with open('../VLGuard/data/forget_data.json', 'r') as fout:
    retain_data = json.load(fout)
    reformulate_data = []
    for sample in retain_data:
        reformulate_data.append(
            {
                'id': f"{sample['id']}",
                'image': sample['image'],
                'conversations': [
                    {'from': 'human', 'value': sample['instr-resp'][0]['instruction']},
                    {'from': 'gpt', 'value': sample['instr-resp'][0]['response']}
                ]
            }
        )
with open('../VLGuard/data/forget_data.json', 'w') as fin:
    json.dump(reformulate_data, fin, indent=4)
```
